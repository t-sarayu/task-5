# **Exploratory Data Analysis (EDA)**

```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading data
data = pd.read_csv('/content/Titanic.csv')
data
```

Using - .describe(), .info(), .value_counts()


```python
data.describe()
```

```python
data.info()
```

```python
categorical_cols = data.select_dtypes(include=['object']).columns
for col in categorical_cols:
    print(f"Column: {col}")
    print(data[col].value_counts())
    print()
```

Clean Data for Visuals

```python
# Drop rows with missing Age and Embarked for cleaner plots
data_clean=data.dropna(subset=['Age', 'Embarked'])
data_clean
```

Using - sns.pairplot(), sns.heatmap() for visualization


```python
sns.pairplot(data_clean[['Pclass','Age','SibSp','Parch','Fare','Survived']], hue='Survived', vars=['Pclass','Age','SibSp','Parch','Fare'])
plt.suptitle('Pairplot of Numerical Features colored by Survival',y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(data_clean.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Histogram of Age
plt.figure(figsize=(8, 5))
sns.histplot(data_clean['Age'], bins=30, kde=True, color='skyblue')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
#Boxplot: Fare vs Survived
plt.figure(figsize=(8, 5))
sns.boxplot(x='Survived', y='Fare', data=data_clean)
plt.title("Fare vs Survival")
plt.xlabel("Survived")
plt.ylabel("Fare")
plt.show()
```

```python
#Scatterplot: Age vs Fare (hue=Survived)
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=data_clean)
plt.title("Age vs Fare by Survival")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

# Observation:
* Higher Fare correlates with higher survival.

*  First class passengers (low Pclass) are more likely to survive.
* Most passengers were young adults.
* Males had lower survival rate.
